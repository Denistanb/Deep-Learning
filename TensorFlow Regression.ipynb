```python
# Importing

import tensorflow as tf
```

### Creating Custom Data

```python
# Creating X and y Tensors

import numpy as np
X=tf.constant(np.arange(-20,24,4))
y=X+10
X,y
```

```python
# Visualising the X and y

import matplotlib.pyplot as plt
plt.plot(X,y);
plt.xlabel("Input");
plt.ylabel("Output");
plt.title("Square of a Number");
```

### Modeling and Predicting

```python
# Creating a Model and Fitting the Data into It

model=tf.keras.Sequential([
    tf.keras.layers.Dense(1,input_shape=[1]) # Output Layer (Parameter with Input Layer)
])

model.compile(loss="mae",optimizer="SGD",metrics=["mae"])
model.summary()
model.fit(X,y,epochs=100)
```

```python
# Predicting Using Model

model.predict(tf.constant([24]))
```

### Improving Our Model

```python
# Improving

model=tf.keras.Sequential([
    tf.keras.layers.Dense(100,activation="relu",input_shape=[1]),
    tf.keras.layers.Dense(100,activation="relu",input_shape=[1]),
    tf.keras.layers.Dense(100,activation="relu",input_shape=[1]),
    tf.keras.layers.Dense(1,input_shape=[1])
])

model.compile(loss="mae",optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),metrics=["mae"])
model.summary()
model.fit(X,y,epochs=200)
```

```python
model.predict(tf.constant([24]))
```

### Creating a Large Custom Dataset

```python
X=tf.range(-100,101,4)
y=X+10
X,y
```

```python
# Splitting the Data

X_train,X_test,y_train,y_test=X[:40],X[40:],y[:40],y[40:]
```

```python
# Visualising the Data

plt.figure(figsize=(10,6));
plt.scatter(X_train,y_train,c="b",label="Training Data");
plt.scatter(X_test,y_test,c="g",label="Testing Data");
plt.legend();
```

### Training the Data

```python
# Creating a Model and Fitting the Data into It

model=tf.keras.Sequential([
    tf.keras.layers.Dense(100,activation="relu",input_shape=[1]),
    tf.keras.layers.Dense(100,activation="relu",input_shape=[1]),
    tf.keras.layers.Dense(100,activation="relu",input_shape=[1]),
    tf.keras.layers.Dense(1,input_shape=[1])
])
model.compile(loss="mae",optimizer=tf.keras.optimizers.Adam(0.01),metrics=["mae"])
model.summary()
model.fit(X_train,y_train,epochs=200,verbose=1)
```

```python
# Visualising the Model

from tensorflow.keras.utils import plot_model
plot_model(model=model,show_shapes=True)
```

### Predicting the Data

```python
# Prediciting

y_pred=model.predict(X_test)
y_pred
```

```python
# Visulising the Predictions

def plot_pred(training_data=X_train,training_label=y_train,test_data=X_test,test_label=y_test,predictions=y_pred):
  plt.figure(figsize=(10,6));
  plt.scatter(training_data,training_label,c="b",label="Training Data");
  plt.scatter(test_data,test_label,c="g",label="Testing Data");
  plt.scatter(test_data,predictions,c="r",label="Predictions Data");
  plt.legend();
plot_pred()
```

### Evaluating the Model

```python
y_test,tf.squeeze(y_pred)
```

```python
mae=tf.keras.losses.MAE(y_test,tf.squeeze(y_pred))
mse=tf.keras.losses.MSE(y_test,tf.squeeze(y_pred))
mae,mse
```

### Experimenting using Creating Models

```python
# Creating a Model 1

model1=tf.keras.Sequential([
    tf.keras.layers.Dense(1,input_shape=[1])
])
model1.compile(loss="mae",optimizer=tf.keras.optimizers.Adam(0.01),metrics=["mae"])
model1.summary()
model1.fit(X_train,y_train,epochs=100,verbose=1)
```

```python
# Predictions on Model 1 and Visualize it

y_pred1=model1.predict(X_test)
plot_pred(predictions=y_pred1)
```

```python
# Creating a Model 2

model2=tf.keras.Sequential([
    tf.keras.layers.Dense(100,input_shape=[1]),
    tf.keras.layers.Dense(1,input_shape=[1])
])
model2.compile(loss="mae",optimizer=tf.keras.optimizers.Adam(0.01),metrics=["mae"])
model2.summary()
model2.fit(X_train,y_train,epochs=100,verbose=1)
```

```python
# Predictions on Model 2 and Visualize it

y_pred2=model2.predict(X_test)
plot_pred(predictions=y_pred2)
```

```python
# Creating a Model 3

model3=tf.keras.Sequential([
    tf.keras.layers.Dense(100,input_shape=[1]),
    tf.keras.layers.Dense(1,input_shape=[1])
])
model3.compile(loss="mae",optimizer=tf.keras.optimizers.Adam(0.01),metrics=["mae"])
model3.summary()
model3.fit(X_train,y_train,epochs=500,verbose=1)
```

```python
# Predictions on Model 3 and Visualize it

y_pred3=model3.predict(X_test)
plot_pred(predictions=y_pred3)
```

### Saving and Loading Models

```python
# Saving

model3.save("best_model.h5")
```

```python
# Loading

best_model=tf.keras.models.load_model("/content/best_model.h5",custom_objects={"mae": tf.keras.metrics.MeanAbsoluteError()}) # custom_objects - Used to Map the any Custom Classes that are used in this Models
y_predb=best_model.predict(X_test)
plot_pred(predictions=y_predb)
```
