```python
# Importing

import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Creating Dataset

```python
# Creating Circle Dataset

from sklearn.datasets import make_circles
X,y=make_circles(1000,noise=0.03,random_state=30)
X.shape,y.shape
```

```python
# Creating a Dataframe for Storing Dataset

df=pd.DataFrame({"X0" : X[:,0] , "X1" : X[:,1] , "y" : y})
df
```

```python
# Visualising the Data

plt.scatter(X[:,0],X[:,1],c=y,cmap="RdYlBu");
```

### Modelling

```python
# Building and Evaluating the Model - 1

model1=tf.keras.Sequential([
    tf.keras.layers.Dense(1,input_shape=(2,)),
    tf.keras.layers.Dense(1)
])
model1.compile(loss=tf.keras.losses.BinaryCrossentropy(),optimizer=tf.keras.optimizers.SGD(),metrics=["accuracy"])
model1.summary()
model_history1=model1.fit(X,y,epochs=100,verbose=1)
print("Evalating...")
model1.evaluate(X,y)
```

```python
# Building and Evaluating the Model - 2

model2=tf.keras.Sequential([
    tf.keras.layers.Dense(100,input_shape=(2,)),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])
model2.compile(loss=tf.keras.losses.BinaryCrossentropy(),optimizer=tf.keras.optimizers.Adam(),metrics=["accuracy"])
model2.summary()
model_history2=model2.fit(X,y,epochs=100,verbose=1)
print("Evalating...")
model2.evaluate(X,y)
```

### Visualising

```python
 # Plotting Decision Boundary

def plot_boundary(model,X,y):
  # Making Boundaries
  x_min,x_max=X[:,0].min()-0.1,X[:,0].max()+0.1
  y_min,y_max=X[:,1].min()-0.1,X[:,1].max()+0.1
  xx,yy=np.meshgrid(np.linspace(x_min,x_max,100),np.linspace(y_min,y_max,100))

  # Creating X Values
  x_in=np.c_[xx.ravel(),yy.ravel()]

  # Making Predictions
  y_pred=model.predict(x_in)

  # Extracting the Output
  if(y_pred[0]>1): # Multi Class
    y_pred1=np.argmax(y_pred,axis=1).reshape(xx.shape)
  else: # Binary Class
    y_pred1=np.round(y_pred).reshape(xx.shape)

  # Plotting the Decision Boundary
  plt.contourf(xx,yy,y_pred1,cmap=plt.cm.RdYlBu,alpha=0.7)
  plt.scatter(X[:,0],X[:,1],c=y,s=40,cmap=plt.cm.RdYlBu)
  plt.xlim(xx.min(),xx.max())
  plt.ylim(yy.min(),yy.max())

plot_boundary(model=model2,X=X,y=y)
```

### Splitting Data into Train and Test Set and Modelling it



```python
# Splitting

X_train=X[:800]
X_test=X[800:]
y_train=y[:800]
y_test=y[800:]

# Building the Model and Fit it

model5=tf.keras.Sequential([
    tf.keras.layers.Dense(4,input_shape=(2,),activation="relu"),
    tf.keras.layers.Dense(4,activation="relu"),
    tf.keras.layers.Dense(1,activation="sigmoid")
])
model5.compile(loss=tf.keras.losses.BinaryCrossentropy(),optimizer=tf.keras.optimizers.Adam(0.01),metrics=["accuracy"])
model5.summary()
model_history5=model5.fit(X_train,y_train,epochs=100,verbose=1)
print("Evalating...")
model5.evaluate(X_test,y_test)
```

```python
# Plotting the Decision Boundary for Training and Testing Set

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.title("Training Set")
plot_boundary(model5,X_train,y_train)
plt.subplot(1,2,2)
plt.title("Testing Set")
plot_boundary(model5,X_test,y_test)
```

```python
# Plotting the Training Performance of a Model

plt.title("Model Performance");
plt.xlabel("Epochs");
plt.ylabel("Accuracy and Loss");
model_df=pd.DataFrame(model_history5.history)
plt.plot(model_df);
```

### Using a Callback - Learning Rate Scheduler

```python
# Building the Model and Fit it

model6=tf.keras.Sequential([
    tf.keras.layers.Dense(4,input_shape=(2,),activation="relu"),
    tf.keras.layers.Dense(4,activation="relu"),
    tf.keras.layers.Dense(1,activation="sigmoid")
])
model6.compile(loss=tf.keras.losses.BinaryCrossentropy(),optimizer=tf.keras.optimizers.Adam(0.01),metrics=["accuracy"])
model6.summary()
lr_sch=tf.keras.callbacks.LearningRateScheduler(lambda epoch : 1e-4 * 10**(epoch/20)) # Creating a Callback
model_history6=model6.fit(X_train,y_train,epochs=100,verbose=1,callbacks=[lr_sch])
print("Evalating...")
model6.evaluate(X_test,y_test)
```

```python
# Plotting the Learning Rate Vs. Loss

model_df2=pd.DataFrame(model_history6.history)
lrs=1e-4 * (10**(tf.range(100)/20))
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.title("Learning Rate Vs. Loss");
plt.xlabel("Learning Rate");
plt.ylabel("Loss");
plt.semilogx(lrs,model_df2["loss"]);
plt.subplot(1,2,2)
plt.title("Learning Rate Vs. Accuracy")
plt.xlabel("Learning Rate");
plt.ylabel("Accuracy");
plt.semilogx(lrs,model_df2["accuracy"]);
```

### Building a Model with Ideal Learning Rate

```python
# Building the Model - 7

model7=tf.keras.Sequential([
    tf.keras.layers.Dense(4,input_shape=(2,),activation="relu"),
    tf.keras.layers.Dense(4,activation="relu"),
    tf.keras.layers.Dense(1,activation="sigmoid")
])
model7.compile(loss=tf.keras.losses.BinaryCrossentropy(),optimizer=tf.keras.optimizers.Adam(0.02),metrics=["accuracy"])
model7.summary()
model_history7=model7.fit(X_train,y_train,epochs=20,verbose=1)
print("Evalating...")
model7.evaluate(X_test,y_test)
```

### Evaluating the Model

```python
# Evaluation Metrics

y_pred=tf.cast(tf.round(model7.predict(X_test)),tf.int64)

acc=tf.keras.metrics.Accuracy()
acc.update_state(y_test, y_pred)
acc_sco=acc.result().numpy()

pre=tf.keras.metrics.Precision()
pre.update_state(y_test, y_pred)
pre_sco=acc.result().numpy()

re=tf.keras.metrics.Recall()
re.update_state(y_test, y_pred)
re_sco=acc.result().numpy()

f1=tf.keras.metrics.F1Score()
f1.update_state(tf.expand_dims(y_test, axis=-1) , y_pred)
f1_sco=acc.result().numpy()

print(f"Accuracy: {acc_sco}, Precision: {pre_sco}, Recall: {re_sco}, F1-Score: {f1_sco}")
```

```python
# Plotting Confusion Matrix

import itertools
from sklearn.metrics import confusion_matrix

def plot_con_mat(y_test,y_pred):
  con_mat=confusion_matrix(y_test,y_pred)
  con_mat_norm=con_mat/sum(con_mat) # Normalising the Matrix
  n_classes=con_mat.shape[0]

  # Creating Plots

  fig,ax=plt.subplots(figsize=(10,10))
  cax=ax.matshow(con_mat,cmap=plt.cm.Blues) # Matrix Plot
  fig.colorbar(cax) # Creating Color Bar
  classes=False
  if classes:
    labels=classes # Multi Class
  else:
    labels=np.arange(con_mat.shape[0]) # Binary Class
  ax.set(title="Confusion Matrix" , xlabel="Predicted Label" , ylabel="True Label" , xticks=np.arange(n_classes) , yticks=np.arange(n_classes) , xticklabels=labels , yticklabels=labels); # Adding Components in Plot
  ax.xaxis.set_label_position("bottom"); # Label is Placed at Bottom
  ax.xaxis.tick_bottom(); # Creating Ticks at X-axis
  ax.xaxis.label.set_size(20) # Adjusting Label Size
  ax.yaxis.label.set_size(20) # Adjusting Label Size
  ax.title.set_size(20) # Adjusting Title Size
  threshold=(con_mat.max()+con_mat.min())/2. # Threshold is 50
  for i,j in itertools.product(range(con_mat.shape[0]),range(con_mat.shape[0])): # 2 -> 00 01 10 11
    plt.text(j,i,f"{con_mat[i,j]} ({con_mat_norm[i,j]*100:.1f}%)",horizontalalignment="center",color="white" if con_mat[i,j]>threshold else "black",size=15) # Adding Text to Each Box

plot_con_mat(y_test,y_pred)
```
